#Imports de librerias necesarias


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import sqlite3
```

# 1 importar el dataset con pandas


```python

df = pd.read_csv('ventas.csv')
df
```

# 2 exploracion de los datos


```python
print(f'1️⃣primeras filas del dataframe :{df.head()} ')
print(f'2️⃣ultimas filas del dataframe :{df.tail()} ')
print(f'3️⃣ info general del dataframe :{df.info} ')
print(f'4️⃣ tipos de dato por columna del dataframe : {df.dtypes} ')
print(f'5️⃣ dimensiones: {df.shape} ')
print(f'6️⃣ dimensiones: {df.describe} ')
print(f'7️⃣ valores nulos por columna: {df.isnull().sum()} ')
```

# 3 limpieza de los datos


```python

#rellenar nulos de producto con 'desconocido'
df['Producto'] = df['Producto'].fillna('Desconocido')

#obtener mediana del precio para rellenar en registros nulos
mediana=df['Precio_Unitario'].median()
print(f'mediana: {mediana}')
df['Precio_Unitario']=df['Precio_Unitario'].fillna(mediana)

#limpiar pago
df['Metodo_Pago']=df['Metodo_Pago'].fillna('No especifica')

#limpiar region
df['Region']=df['Region'].fillna('No especifica')

#se calculan los totales de las ventas a partir del precio unitario y cantidad
#los nulos se rellenan con la mediana
df['Total_Venta'] = df['Cantidad'] * df['Precio_Unitario']
mediana_ventas=df['Total_Venta'].median()
df['Total_Venta'].fillna(mediana_ventas)


#se formatean las fechas
#hay varios formatos de fecha que unificar
fecha_formato1=pd.to_datetime(df['Fecha'], format='%d-%m-%Y', errors='coerce')
fecha_formato2=pd.to_datetime(df['Fecha'], format='%Y-%m-%d', errors='coerce')

df['Fecha']=fecha_formato1.combine_first(fecha_formato2)


#columnas textuales capitalizadas
df['Nombre_Cliente']=df['Nombre_Cliente'].str.title()
df['Apellido_Cliente']=df['Apellido_Cliente'].str.title()
df['Producto']=df['Producto'].str.title()
df['Metodo_Pago']=df['Metodo_Pago'].str.title()
df['Region']=df['Region'].str.title()


#POST LIMPIEZA CORROBORAMOS
print(df.head())

print(f'🔰datos nulos post limpieza: {df.isnull().sum()}')
```

# 4 Estadistica descriptiva


```python
desc=df.describe(include='all')
print(f'1️⃣ estadisticas: {desc}')


print('2️⃣estadistica de total venta x region')
print(df.groupby('Region')['Total_Venta'].describe(include='all'))
```

# 5 Graficos


```python
#histograma (total ventas)
plt.figure(figsize=(8,5))
#se usa dropna solo por si acaso para no romper el grafico, pero a esta altura los datos ya se encuentran limpios
plt.hist(df['Total_Venta'].dropna(),bins=15,edgecolor='black',color='skyblue')
plt.title('Distribucion de ventas')
plt.xlabel('Total venta')
plt.ylabel('Frecuencia')
plt.grid(axis='y',linestyle='--')
plt.show()


#grafico de barras (ventas x region)
regiones = df['Region'].value_counts()
plt.figure(figsize=(8,5))
plt.bar(regiones.index,regiones.values,)
plt.title('Numero de ventas por region')
plt.xlabel('Region')
plt.ylabel('Numero de ventas')
plt.grid(axis='y',linestyle='--')
plt.show()

```

# 6 Exportar dataframe a sql con sqlite


```python
db_nombre='db_procesada.db'
nombre_tabla='ventas_datos_limpios'

conn=sqlite3.connect(db_nombre)
df.to_sql(nombre_tabla,conn,if_exists='replace',index=False)
conn.close()

print('✔ dataframe exportado con exito a sql')
```
